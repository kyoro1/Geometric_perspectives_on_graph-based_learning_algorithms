```python
from common import EdgePruningUtil
```

```python
ep = EdgePruningUtil()
```

```python
## Grid search over hyperparameters through parallel jobs
DATASETS = ['concentric_circles', 'moons', 's_curve', 'cassini', 'mixture_of_gaussians']
DATASETS_2DIM = ['torii', 'hyperboloids', 'parab_and_hyp', 'double_paraboloid', '3D_swiss_roll']
DELTAS = [0.70, 0.75, 0.8, 0.85, 0.90]
LAMS   = [0.001, 0.01, 0.1, 0.2, 0.5]

util = EdgePruningUtil()
df_result = util.grid_search_joblib(
    deltas=DELTAS,
    lams=LAMS,
    datasets=DATASETS,
    datasets_2dim=DATASETS_2DIM,
    n_points=4000,
    n_jobs=None,              # defaults to CPU cores - 1
    pre_dispatch="2*n_jobs",  # limit tasks in-flight to control memory
    batch_size="auto",
    verbose=10,
    need_data=False,
)
```
