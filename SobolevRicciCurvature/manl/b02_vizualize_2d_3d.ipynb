```python
import joblib
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.graph_objects as go
from viz_demo import Visualization

import matplotlib as mpl
mpl.rcParams["text.usetex"] = False
```

```python
def load_datapoints(datasets: list,
                    datasets_2dim: list,
                    n_points: int,) -> list:
    '''
    Load precomputed data points and their associated information from disk.
    So, please make sure that the data files are available in './data/' directory.
    Parameters:
    - datasets: List of dataset names for 2D datasets.
    - datasets_2dim: List of dataset names for 3D datasets.
    - n_points: Number of data points in each dataset.

    Returns:
    - je_tmp: List of loaded data objects.
    '''
    je_tmp = []
    for dataset in datasets:
        je = joblib.load(f'./data/je_{dataset}_n{n_points}_2dFalse.pkl')
        je_tmp.append(je)
    for dataset in datasets_2dim:
        je = joblib.load(f'./data/je_{dataset}_n{n_points}_2dTrue.pkl')
        je_tmp.append(je)
    return je_tmp
```

```python
datasets = ['concentric_circles', 'moons', 's_curve', 'cassini', 'mixture_of_gaussians']
datasets_2dim = ['torii', 'hyperboloids', 'parab_and_hyp', 'double_paraboloid', '3D_swiss_roll']

je_tmp = load_datapoints(datasets, datasets_2dim, n_points=4000)
```

```python
X_list = [je.data for je in je_tmp[:5]]
G_list = [je.G for je in je_tmp[:5]]
shortcut_edges_list = [je.shortcut_edges for je in je_tmp[:5]]
```

```python
vi = Visualization()
fig = vi.plot_graphs_2D_side_by_side(X_list=X_list, 
                                  G_list=G_list, 
                                  shortcut_edges_list=shortcut_edges_list,
                                  shortcut_width=2,
                                  subtitle_font_size=35,
                                  titles=['concentric_circles', 'moons', 's_curve', 'cassini', 'mixture_of_gaussians'],
                                  )
```

## `Figure 11`

```python
#datasets_2dim = ['torii', 'hyperboloids', 'parab_and_hyp', 'double_paraboloid', '3D_swiss_roll']
datasets_2dim = '3D_swiss_roll'
n_points = 4000
two_dim = True

je = joblib.load(f'./data/je_{datasets_2dim}_n{n_points}_2d{two_dim}.pkl')
```

```python
## Calculate ORC and SC(SPT)
orc_kappa, src_kappa = je.calc_curvatures(sc_spt_kNN=20)
```

```python
## Extract non-zero curvatures
orc = orc_kappa[orc_kappa != 0]
sc = src_kappa[src_kappa != 0]
```

```python
plt.figure(figsize=(8,6))
sns.histplot(orc, kde=False, label="ORC", bins=80)
sns.histplot(sc, kde=False, label="SRC(SPT)", bins=120)
plt.legend(fontsize=15)
plt.xlim(-1.0, 1.0)
plt.xlabel("Curvature values", fontsize=15)
plt.savefig("./results/orc_sc_hist.pdf")
```
