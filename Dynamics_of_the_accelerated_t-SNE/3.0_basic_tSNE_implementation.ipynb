```python
import utils
import seaborn as sns
```

```python
## Basic configuration
sample_data = 'kddcup'
h = 10
perplexity = 30
alpha = 10
labels = ['smurf', 'neptune', 'normal', 'back', 'satan']
total_iteration = 50
sample_number = 100
momentum = 0.5
R = 5

exp = utils.tSNE_experiment(labels=labels, 
                           sample_data=sample_data, 
                           h=h, 
                           perplexity=perplexity, 
                           alpha=alpha,
                           total_iteration=total_iteration,
                           sample_number=sample_number,
                           momentum=momentum,
                           R=R)
```

```python
## Load KDDCup data
exp.load_data()
```

```python
# Calculate the probability matrix P
exp.tsne_probabilities(X=exp.X_sampled, 
                    perplexity=30,
                    tol=1e-10,
                    initial_beta_coefficient=1e-4)
```

```python
## Heatmap of ajacency matrix P
exp.show_P_matrix()
```

```python
## Initialize Y (embedding in 2-dim.)
exp.initialize_Y()
```

```python
## Optimization
# exp.calculate_YQS(optimization_method='GD')  # For GD
# exp.calculate_YQS(optimization_method='MM') # For MM
exp.calculate_YQS(optimization_method='NAG') # For NAG
```

```python
## Visualization of the result
iteration = exp.total_iteration - 1

sns.scatterplot(x=exp.Y[iteration][:,0], y=exp.Y[iteration][:,1], hue=exp.y_sampled, palette="deep")
```
